##Parte a:

Primero definimos las funciones asociadas a las 2 ecuaciones diferenciales:

```python
def F1(a,m,k1): 
  ec1 = 1 - m - a - (k1 * m**2 * a) 
  return ec1

def F2(a,m,alfa,k1,k2): 
  ec2 = (k1 * m**2 * a) - (k2 * m**alfa) * (1 - m - a) 
  return ec2
```

Ahora creamos una funcion que resuelva el sitema completo:

```python
#sistema: int x 7 -> [array x4]
#pide como parametros: el largo del intervalo (L), un paso (h), k1, k2,
#alfa, a0 y m0 (condiciones iniciales)
#luego devuelve una lista con 4 arreglos: t, a(t), m(t) y s(t)
def sistema(L,h,k1,k2,alfa,a0,m0):
    
    n = int(L/h)  #cantidad de iteraciones

    #definimos inicialmente t, a, m con ceros:
    
    t = np.zeros(n)
    a = np.zeros(n)
    m = np.zeros(n)
    
    #definimos las condiciones iniciales:
    
    t[0] = 0
    a[0] = a0
    m[0] = m0
    
    #calculamos a(t) y m(t) usando el metodo de euler progresivo:
        
    for i in np.arange(1,n):
    
        a[i] = a[i-1] + h * F1(a[i-1],m[i-1],k1)
        m[i] = m[i-1] + h * F2(a[i-1],m[i-1],alfa,k1,k2)
        t[i] = t[i-1] + h
    
    #calculamos s(t) sabiendo que la suma de a, m y t debe ser igual a 1:
        
    s = np.zeros(n)  #definimos inicialmente con ceros
    
    for j in np.arange(0,n):
        s[j] = 1 - a[j] - m[j]
    
    #ahora retornamos en una lista los 4 arreglos:
    return [t,a,m,s]
```

#Parte c:

La idea sera encontar el periodo limite buscando el tiempo entre 2 maximos consecutivos

```python
#periodo: int x 3 -> int
#pide como parametro un valor de alfa, un paso y un largo de intervalo
#devuelve el periodo limite de s(t)
def periodo(al,h,L):
    
    #primero resolvemos el sistema:
    
    n = int(L/h)  #cantidad de iteraciones
    
    st = sistema(L, h, 8, 15, al, 0.4, 0.3)
    s = st[3]  #s(t)
    
    l = [] #lista que retornara los tiempos donde hay maximos
    
    #buscamos los maximos de sf:
    for i in np.arange(1,n-1):
        
        if s[i-1] < s[i] and s[i] > s[i+1]:
            l.append(i*h)            
            
    #para buscar el periodo limite buscamos la diferencia entre Los 2 ultimos
    #elementos L
    #print(l)
    largo = len(l)
    p = l[largo-1] - l[largo-2]
    return p
```

Calculamos el periodo limite para distintos valores de alpha:

```python
p1 = periodo(1.3, 0.05, 100) #usamos un valor menor para que converja 
p2 = periodo(1.4, 0.1, 500)
p3 = periodo(1.5, 0.1, 500)
p4 = periodo(1.6, 0.1, 500)
p5 = periodo(1.7, 0.1, 500)
p6 = periodo(1.8, 0.1, 100)  #usamos un valor menor para que converja
p7 = periodo(1.9, 0.1, 100)  #usamos un valor menor para que converja
```

Graficamos el periodo en funcion de alpha:

```python
Pe = [p1,p2,p3,p4,p5,p6,p7] #periodos limites en cada caso.
x = [1.3,1.4,1.5,1.6,1.7,1.8,1.9]
plt.plot(x,Pe)
plt.title('Periodo límite en función de alpha (Euler progresivo)') 
plt.xlabel('Alpha')
plt.ylabel('Periodo en millones de años')
plt.grid()
plt.show()
```

#Parte d:

Creamos otra funcion que resuelva el sistema, pero usando Runge-Kutta orden 4:

```python
#RK: int x 7 -> [array x4]
#pide como parametros: el largo del intervalo (L), un paso (h), k1, k2,
#alfa, a0 y m0 (condiciones iniciales)
#luego devuelve una lista con 4 arreglos: el tiempo, a(t), m(t) y s(t)
#resuelve el sistema usando el metodo de Runge-Kutta orden 4
def RK(L,h,k1,k2,alfa,a0,m0):

    n = int(L/h)  #cantidad de iteraciones

    #definimos inicialmente t, a, m con ceros:
    
    t = np.zeros(n)
    a = np.zeros(n)
    m = np.zeros(n)
    
    #definimos las condiciones iniciales:
    
    t[0] = 0
    a[0] = a0
    m[0] = m0
    
    #calculamos a(t) y m(t) usando el metodo de Runge-Kutta orden 4:
        
    for i in np.arange(1,n):
        #ctes
        g1 = F1(a[i-1],m[i-1],k1)
        c1 = F2(a[i-1],m[i-1],alfa,k1,k2)
        g2 = F1(a[i-1]+h*g1/2,m[i-1]+h*c1/2,k1)    
        c2 = F2(a[i-1]+h*g1/2,m[i-1]+h*c1/2,alfa,k1,k2)
        g3 = F1(a[i-1]+h*g2/2,m[i-1]+h*c2/2,k1)
        c3 = F2(a[i-1]+h*g2/2,m[i-1]+h*c2/2,alfa,k1,k2)
        g4 = F1(a[i-1]+g3*h,m[i-1]+c3*h,k1)
        c4 = F2(a[i-1]+g3*h,m[i-1]+c3*h,alfa,k1,k2)
        #soluciones:
        a[i] = a[i-1] + h*(g1 + 2*g2 + 2*g3 + g4)/6
        m[i] = m[i-1] + h*(c1 + 2*c2 + 2*c3 + c4)/6
        t[i] = t[i-1] + h

    #calculamos s(t) sabiendo que la suma de a, m y t debe ser igual a 1:
        
    s = np.zeros(n)
    
    for j in np.arange(0,n):
        s[j] = 1 - a[j] - m[j]
    
    #ahora retornamos en una lista los 4 arreglos:
    return [t,a,m,s]
```

Creamos otra funcion de periodo, pero usando el metodo de Runge-Kutta:

```python
#per: int x 3 -> int
#pide como parametro un valor de alfa, un paso y un largo de intervalo
#devuelve el periodo limite de s(t)
def per(al,h,L):
    
    #primero resolvemos el sistema:
    
    n = int(L/h)  #cantidad de iteraciones
    assert n > 100 
    
    sol = RK(L, h, 8, 15, al, 0.4, 0.3)
    s = sol[3]  #s(t)
    
    l = [] #lista que retornara el tiempo donde hay maximos
    
    #buscamos los maximos de sf:
    for i in np.arange(1,n-1):
        
        if s[i-1] < s[i] and s[i] > s[i+1]:
            l.append(i*h)           
        
    #para buscar el periodo limite buscamos la diferencia entre Los 2 ultimos
    #elementos L
    largo = len(l)
    p = l[largo-1] - l[largo-2] #ojo
    return p
```

Calculamos el periodo limite para distintos valores de alpha:

```python
p1 = per(1.3, 0.05, 100) #usamos un valor menor para que converja
p2 = per(1.4, 0.1, 500)
p3 = per(1.5, 0.1, 500)
p4 = per(1.6, 0.1, 500)
p5 = per(1.7, 0.1, 500)
p6 = per(1.8, 0.1, 100) #usamos un valor menor para que converja
p7 = per(1.9, 0.1, 100) #usamos un valor menor para que converja
```

Graficamos el periodo en funcion de alpha:

```python
Prk = [p1,p2,p3,p4,p5,p6,p7] #periodos limites en cada caso.
x = [1.3,1.4,1.5,1.6,1.7,1.8,1.9]
plt.plot(x,Prk)
plt.figtext(.5,.9,'Periodo límite en función de alpha (Runge Kutta)', fontsize=20, ha='center') 
plt.xlabel('Alpha')
plt.ylabel('Periodo en millones de años')
plt.grid()
plt.show()
```

#Parte e:

Resolvemos el sistema usando la libreria scipy.integrate.odeint. Se usa la cuadratura de Gauss–Kronrod para aproximar la integral.

Primero debemos definir las ecuaciones del sistema como una funcion.Lamentablemente, no podemos resolver el sistema dejando alpha como variable al usar odeint, luego definimos las ecuaciones para cada valor de alpha:

```python
def ecuaciones1(x,t):
    #constantes:
    k1 = 8
    k2 = 15
    alpha = 1.3
    #se asigna cada ecuacion con un elemento del vector
    A = x[0]
    M = x[1]
    #S = x[2]
    #definimos las ecuaciones:
    dadt = 1 - M - A - (k1 * M**2 * A)
    dmdt = (k1 * M**2 * A) - (k2 * M**alpha) * (1 - M - A)
    return [dadt, dmdt]

def ecuaciones2(x,t):
    #constantes:
    k1 = 8
    k2 = 15
    alpha = 1.4
    #se asigna cada ecuacion con un elemento del vector
    A = x[0]
    M = x[1]
    #S = x[2]
    #definimos las ecuaciones:
    dadt = 1 - M - A - (k1 * M**2 * A)
    dmdt = (k1 * M**2 * A) - (k2 * M**alpha) * (1 - M - A)
    return [dadt, dmdt]

def ecuaciones3(x,t):
    #constantes:
    k1 = 8
    k2 = 15
    alpha = 1.5
    #se asigna cada ecuacion con un elemento del vector
    A = x[0]
    M = x[1]
    #S = x[2]
    #definimos las ecuaciones:
    dadt = 1 - M - A - (k1 * M**2 * A)
    dmdt = (k1 * M**2 * A) - (k2 * M**alpha) * (1 - M - A)
    return [dadt, dmdt]

def ecuaciones4(x,t):
    #constantes:
    k1 = 8
    k2 = 15
    alpha = 1.6
    #se asigna cada ecuacion con un elemento del vector
    A = x[0]
    M = x[1]
    #S = x[2]
    #definimos las ecuaciones:
    dadt = 1 - M - A - (k1 * M**2 * A)
    dmdt = (k1 * M**2 * A) - (k2 * M**alpha) * (1 - M - A)
    return [dadt, dmdt]

def ecuaciones5(x,t):
    #constantes:
    k1 = 8
    k2 = 15
    alpha = 1.7
    #se asigna cada ecuacion con un elemento del vector
    A = x[0]
    M = x[1]
    #S = x[2]
    #definimos las ecuaciones:
    dadt = 1 - M - A - (k1 * M**2 * A)
    dmdt = (k1 * M**2 * A) - (k2 * M**alpha) * (1 - M - A)
    return [dadt, dmdt]

def ecuaciones6(x,t):
    #constantes:
    k1 = 8
    k2 = 15
    alpha = 1.8
    #se asigna cada ecuacion con un elemento del vector
    A = x[0]
    M = x[1]
    #S = x[2]
    #definimos las ecuaciones:
    dadt = 1 - M - A - (k1 * M**2 * A)
    dmdt = (k1 * M**2 * A) - (k2 * M**alpha) * (1 - M - A)
    return [dadt, dmdt]

def ecuaciones7(x,t):
    #constantes:
    k1 = 8
    k2 = 15
    alpha = 1.9
    #se asigna cada ecuacion con un elemento del vector
    A = x[0]
    M = x[1]
    #S = x[2]
    #definimos las ecuaciones:
    dadt = 1 - M - A - (k1 * M**2 * A)
    dmdt = (k1 * M**2 * A) - (k2 * M**alpha) * (1 - M - A)
    return [dadt, dmdt]
```

Definimos una función que busque el periodo límite:

```python
#limite: int x 3 -> int
#pide como parametro un valor de alfa, un paso y un largo de intervalo
#devuelve el periodo limite de s(t)
def p_limite(alpha,h,l):
    
    n = int(l/h)  #cantidad de iteraciones
    
    #primero resolvemos el sistema:
        
    #definimos las condiciones iniciales:
    x0 = [0.4, 0.3]
    
    #definimos t:
    t = np.linspace(0,l,n)
    
    #definimos una matriz identidad que nos servira para resolver el sistema
    x = np.identity(l)
    
    #resolvemos utilizando odeint para cada caso:
    if alpha == 1.3:
        x = odeint(ecuaciones1,x0,t)
        
    elif alpha == 1.4:
        x = odeint(ecuaciones2,x0,t)
        
    elif alpha == 1.5:
        x = odeint(ecuaciones3,x0,t)
        
    elif alpha == 1.6:
        x = odeint(ecuaciones4,x0,t)
        
    elif alpha == 1.7:
        x = odeint(ecuaciones5,x0,t)
        
    elif alpha == 1.8:
        x = odeint(ecuaciones6,x0,t)
        
    elif alpha == 1.9:
        x = odeint(ecuaciones7,x0,t)

        
    #obtenemos A y M de las columnas de x:
    A = x[:,0]
    M = x[:,1]
    
    #obtenemos S:
    S = np.zeros(n)  #definimos inicialmente con ceros
    
    for j in np.arange(0,n):
        S[j] = 1 - A[j] - M[j]
    
    #ahora obtenemos el periodo:    
    
    L = [] #lista que retornara el tiempo donde hay maximos
    
    #buscamos los maximos de sf:
    for i in np.arange(0,n-1):
        
        if S[i-1] < S[i] and S[i] > S[i+1]:
            L.append(i*h)            
        
    #para buscar el periodo limite buscamos la diferencia entre Los 2 ultimos
    #elementos L
    largo = len(L)
    p = L[largo-1] - L[largo-2]
    return p
```

Calculamos para distintos valores de alpha:

```python
p1 = p_limite(1.3, 0.05, 100) #usamos un valor menor para que converja
p2 = p_limite(1.4, 0.1, 500)
p3 = p_limite(1.5, 0.1, 500)
p4 = p_limite(1.6, 0.1, 500)
p5 = p_limite(1.7, 0.1, 500)
p6 = p_limite(1.8, 0.1, 100) #usamos un valor menor para que converja
p7 = p_limite(1.9, 0.1, 100) #usamos un valor menor para que converja
```

Graficamos el periodo en funcion de alpha:

```python
Po = [p1,p2,p3,p4,p5,p6,p7] #periodos limites en cada caso.
xa = [1.3,1.4,1.5,1.6,1.7,1.8,1.9]
plt.plot(xa,Po)
plt.title('Periodo límite en función de alpha (odeint)') 
plt.xlabel('Alpha')
plt.ylabel('Periodo en millones de años')
plt.grid()
plt.show()
```
